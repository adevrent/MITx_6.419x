```python
import numpy as np
```

```python
def ReLU(z):
    return np.maximum(z, 0)

def calc_s(s0, x, W_ss, W_sx):
    return ReLU(W_ss @ s0 + W_sx @ x)

def calc_y(s, W_sy, W0):
    return np.sign(W_sy @ s + W0)
```

```python
s0 = np.array([0, 0])

W_ss = np.array([
    [-1, 0],
    [0, 1]
])

W_sx = np.array([
    [1, 0],
    [0, 1]
])

AA = np.array([
    [1, 0],
    [1, 0]
]).T

ABB = np.array([
    [1, 0],
    [0, 1],
    [0, 1]
]).T

BAA = np.array([
    [0, 1],
    [1, 0],
    [1, 0]
]).T

print("AA:\n", AA)
print("ABB:\n", ABB)
print("BAA:\n", BAA)
```

```python
calc_s(s0, AA, W_ss, W_sx)
```
