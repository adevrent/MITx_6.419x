```python
import numpy as np
import matplotlib.pyplot as plt

def f(z):  # Activation function for hidden layer
    return np.tanh(z)

def z(w, x, w0):
    # print("z =", np.dot(w, x) + w0)
    return np.dot(w, x) + w0

def calc_a_vector(x, w1, w01, w2, w02):
    a1 = f(z(w1, x, w01))
    a2 = f(z(w2, x, w02))
    return (a1, a2)

X = np.array([
    [-1, -1],
    [1, -1],
    [-1, 1],
    [1, 1]
])

Y = np.array([1, -1, -1, 1])

w1 = np.array([1, -1])
w01 = 1
w2 = np.array([-1, 1])
w02 = 1

a_array = []
for i in range(X.shape[0]): # iterate over rows
    a_array.append(calc_a_vector(X[i], w1, w01, w2, w02))
print("a_array =,", a_array)
plt.scatter([pair[0] for pair in a_array], [pair[1] for pair in a_array])
```
