```python
import numpy as np

# Define the (2, 3) array
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

# Define the (3,) array
scalars = np.array([2, 3, 4])

# Reshape the scalars array to (1, 3) to align with the columns of the matrix
scalars = scalars.reshape(1, 3)  # Equivalent to scalars.reshape(1, 3)

# Multiply each column of the matrix by the corresponding scalar
result = matrix * scalars

print(result)
```

```python
temp_arr = np.array([
    [2, 2, 2],
    [1, 1, 1]
])

long_arr = np.array([1, 1, 1])

temp_arr * long_arr
```
