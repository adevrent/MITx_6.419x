```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [8, 4, 7],
    [2, 8, 1],
    [3, 1, 1],
    [9, 7, 4]
])

print(X)
```

```python
Xbar = X.mean(axis=0)
print(Xbar)
print(Xbar.shape)
```

```python
sum = 0
n = X.shape[0]  # num of rows (num of observations)
for i in range(n):
    sum += np.outer(X[i, :], X[i, :])
sum = sum / n
print("sum =", sum)
S = sum - np.outer(X.mean(axis=0), X.mean(axis=0).T)

print("X.mean(axis=0) @ X.mean(axis=0).T =", X.mean(axis=0) @ X.mean(axis=0).T)

print("sum =", sum)
print("S =\n", S)
```

```python
x1 = np.array([1, 2])
x2 = np.array([3, 4])
x3 = np.array([-1, 0])

u = np.array([1/np.sqrt(5), 2/np.sqrt(5)])

proj_x1 = np.dot(x1, u)*u
proj_x2 = np.dot(x2, u)*u
proj_x3 = np.dot(x3, u)*u

print(proj_x1)
print(proj_x2)
print(proj_x3)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define vectors
vectors = np.array([x1, x2, x3, u])
proj_vectors = np.array([proj_x1, proj_x2, proj_x3, u])

# Define colors
colors = ['r', 'g', 'b', 'k']

# Plot vectors using quiver
fig, axs = plt.subplots(1, 2, figsize=(8, 12))  # Adjust figure size

for i, vector in enumerate(vectors):
    axs[0].quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color=colors[i], label=(f'$x_{i+1}$' if i != len(vectors) - 1 else "$u$"))
    axs[0].set_xlim(-4, 4)
    axs[0].set_ylim(-4, 4)
    axs[0].set_aspect('equal', adjustable='box')
    axs[0].grid(True)
    axs[0].legend(loc="lower right")

for i, vector in enumerate(proj_vectors):
    axs[1].quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color=colors[i], label=(f'$x_{i+1}$' if i != len(proj_vectors) - 1 else "$u$"))
    axs[1].set_xlim(-4, 4)
    axs[1].set_ylim(-4, 4)
    axs[1].set_aspect('equal', adjustable='box')
    axs[1].grid(True)
    axs[1].legend(loc="lower right")

fig.tight_layout()
plt.show()
```
