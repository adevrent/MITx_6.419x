```python
import numpy as np
```

```python
x1 = [-1, -1]
x2 = [1, 0]
x3 = [-1, 1.5]

X_1 = np.array([x1, x2, x3])  # feature vectors as rows
X_1
```

```python
X_2 = np.array([x2, x3, x1])  # feature vectors as rows
X_2
```

```python
y_1 = np.array([1, -1, 1])  # labels
y_1
```

```python
y_2 = np.array([-1, 1, 1])  # labels
y_2
```

```python
def perceptron(X, y, T=20, theta=None):
    n = X.shape[0]
    p = X.shape[1]
    
    if theta is None:
        theta = np.zeros(p)
    
    mistakes = 0
    theta_array = []
    for t in range(T):
        print(f"theta = {theta} at the start of Iteration {t+1}")
        for i in range(n):
            if y[i] * np.dot(theta, X[i]) <= 0:
                mistakes += 1
                theta = theta + y[i] * X[i]
                theta_array.append(theta.tolist())
    
    print("num of mistakes =", mistakes)
    print("theta_progression = ", theta_array)
    return theta
```

```python
perceptron(X_1, y_1)
```

```python
perceptron(X_2, y_2)
```

```python
perceptron(np.array([x1, x2, [-1, 10]]), y_1)
```

```python
perceptron(np.array([x2, x1, [-1, 10]]), y_2)
```

## 2. Perceptron Performance


```python
X = np.array([
    [-4, 2],
    [-2, 1],
    [-1, -1],
    [2, 2],
    [1, -2]
])

y = np.array([1, 1, -1, -1, -1])
```

```python
def perceptron_with_offset(X, y, T=20, theta=None, theta0=None):
    n = X.shape[0]
    p = X.shape[1]
    
    if theta is None:
        theta = np.zeros(p)
    if theta0 is None:
        theta0 = 0
    
    mistakes = 0
    theta_array = []
    theta0_array = []
    for t in range(T):
        print(f"theta = {theta} at the start of Iteration {t+1}")
        print(f"theta0 = {theta0} at the start of Iteration {t+1}")
        for i in range(n):
            if y[i] * (np.dot(theta, X[i]) + theta0) <= 0:
                mistakes += 1
                theta = theta + y[i] * X[i]
                theta0 = theta0 + y[i]
                theta_array.append(theta.tolist())
                theta0_array.append(theta0.tolist())
    
    print("num of mistakes =", mistakes)
    print("theta_progression = ", theta_array)
    print("theta0_progression = ", theta0_array)
    return theta, theta0
```

```python
perceptron_with_offset(X, y, T=20)
```
