```python
X = np.array([[0, 1, 2],
              [0, 1, 2]])
X = np.vstack((np.ones((1, X.shape[1])), X))
print("X =\n", X)

Y = np.array([[0, 1, 0]])

W_T_1 = np.array([
    [-0.5, 1, 0],
    [1.5, -1, 0]
])

def f(z):
    if z > 0:
        return 1
    else:
        return 0
    
Z = W_T_1 @ X
A = np.vectorize(f)(Z)
print("A =\n", A)
# A.tolist()
```

```python
def add_intercept(x):
    """Assumes x is a matrix; augments it with 1 at the beginning

    Args:
        x (np.array): numpy array
    """
    return np.vstack((np.ones(x.shape[1]), x))

W_1 = np.array([1, 1, 1]).reshape((3, 1))
eta = 0.5

x = np.array([1, 2]).reshape((2, 1))
y = -1

z = W_1.T @ add_intercept(x)
z
```

```python
def hinge_loss(w, y, x):
    # Assumes x is augmented col vector (1 at beginning)
    return np.max(0, 1 - y*(w.T @ add_intercept(x)))

def gradient_hinge_loss(w, y, x):
    v = y * (w.T @ add_intercept(x))
    if v > 1:
        return 0
    else:
        return -y * add_intercept(x)
    
def GD_single_step(w, y, x, eta=0.5):
    w = w - eta * gradient_hinge_loss(w, y, x)
    return w

print("W_1 before =", W_1)
W_1 = GD_single_step(W_1, y, x)
print("W_1 after =", W_1)
```
